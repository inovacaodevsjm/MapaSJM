```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Arquivos
arquivo_csv = "../Processed/Saude/PrazoEntregaSaude.csv"
arquivo_geojson = "../Processed/Saude/PrazoEntregaSaude.geojson"

# Nomes das colunas
COL_LAT = "Latitude"
COL_LON = "Longitude"

# Lendo o CSV
df = pd.read_csv(arquivo_csv)

# Criando a geometria (lon, lat ‚Äî essa ordem √© sagrada)
geometry = [
    Point(lon, lat)
    for lon, lat in zip(df[COL_LON], df[COL_LAT])
]

# Criando GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

# Salvando como GeoJSON
gdf.to_file(arquivo_geojson, driver="GeoJSON")

print("CSV convertido para GeoJSON com sucesso üåçüöÄ")
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
# 1. Ler o Excel
df = pd.read_excel("../Processed/Saude/PrazoEntregaSaude.xlsx")  # ajuste o nome do arquivo
```

```python
# 2. Separar latitude e longitude a partir da coluna "Coordenadas"
df[["latitude", "longitude"]] = (
    df["COORDENADAS"]
    .str.split(",", expand=True)
)

# 3. Converter para float (remove espa√ßos automaticamente)
df["latitude"] = df["latitude"].astype(float)
df["longitude"] = df["longitude"].astype(float)

# 4. Criar geometria (Point precisa de longitude primeiro)
df["geometry"] = df.apply(
    lambda row: Point(row["longitude"], row["latitude"]),
    axis=1
)

# 5. Criar GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")
```

```python
# 6. Exportar para GeoJSON
gdf.to_file("../Processed/Saude/PrazoEntregaSaude.geojson", driver="GeoJSON")

print("GeoJSON gerado com sucesso!")
```

```python
#Assistencia Social
import geopandas as gpd

# Arquivos
arquivo_kml = "../Processed/AssistenciaSocial/AssistenciaSocial.kml"
arquivo_geojson = "../Processed/AssistenciaSocial/AssistenciaSocial.geojson"

# Lendo o KML
gdf = gpd.read_file(arquivo_kml, driver="KML")

# Garantindo CRS padr√£o do Google Earth
gdf = gdf.set_crs(epsg=4326, allow_override=True)

# Salvando como GeoJSON
gdf.to_file(arquivo_geojson, driver="GeoJSON")

print("KML convertido para GeoJSON com sucesso üåçüöÄ")
```

```python
import geopandas as gpd
import pandas as pd

# Lendo o GeoJSON
gdf = gpd.read_file("../Processed/AssistenciaSocial/AssistenciaSocial.geojson", driver="GeoJSON")

def parse_description(desc):
    if pd.isna(desc):
        return {}

    partes = desc.split("<br>")
    resultado = {}

    for item in partes:
        if ":" in item:
            chave, valor = item.split(":", 1)
            resultado[chave.strip()] = valor.strip()

    return resultado

# Aplicando o parser
desc_cols = gdf["Description"].apply(parse_description)

# Transformando em DataFrame
desc_df = pd.DataFrame(desc_cols.tolist())

# Unindo ao GeoDataFrame original
gdf = pd.concat([gdf.drop(columns=["Description"]), desc_df], axis=1)

# Salvando GeoJSON final
gdf.to_file("../Processed/AssistenciaSocial/AssistenciaSocial.geojson", driver="GeoJSON")

print("Description normalizada em colunas com sucesso üß†üìä")
```
